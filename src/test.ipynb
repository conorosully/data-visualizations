```python
import numpy as np
import pandas as pd

import folium
from folium.plugins import TimestampedGeoJson
from folium.plugins import HeatMap
```

```python
war_dataset = pd.read_csv("data/Extra-StateWarData_v4.0.csv", encoding = "ISO-8859-1")
war_dataset.head()
```

```python
min_year = min(war_dataset['Year'])
max_year = max(war_dataset['Year'])

time_index = []
data = []

for year in range(min_year,max_year+1):
    time_index.append(year)
    data.append(war_dataset[war_dataset['Year'] == year][['Latitude','Longitude']].values.tolist())
    
weight = 1# default value
for time_entry in data:
    for row in time_entry:
        row.append(weight)
data
```

```python

from datetime import datetime, timedelta


time_index = [
    (datetime.now() + k * timedelta(1)).strftime('%Y-%m-%d') for
    k in range(len(data))
]
```
