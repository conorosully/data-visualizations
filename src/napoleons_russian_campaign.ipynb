```python
import numpy as np
import pandas as pd

import folium
from folium.features import DivIcon
```

```python
data_original = pd.read_csv("data/data.csv")
data_original.head()
```

```python
cities = pd.read_csv("data/cities.csv")
cities
```

```python
data_r = data_original[(data_original['dir'] == 'R')]
data_r
```

```python
temps = pd.read_csv("data/temps.csv")
lat = [55.7,55.15,54.8,54.6,54.3,54.2,54.4,54.3,54.4]
temps['lat'] = lat
temps
```

```python
def addText(loc,city):
    marker = folium.map.Marker(
    loc,
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(10,10),
        html='<div style="font-size: 12pt">{}</div>'.format(city),
        )
    )
    return marker

max_size = data_original['size'][0]
def setWeight(size):
    weight = 60*np.log(1.1+(size/max_size))
    return weight
    

def setColour(direction):
    if direction == 'A':
        return '#4286f4'
    else:
        return '#000000'
```

```python
for i in range(3,0,-1):
    print(i)
```

```python
m = folium.Map([55.2, 30.6],zoom_start=7,min_zoom=7,max_zoom =8)

for d in range(3,0,-1):
    data = data_original[data_original['num']==d]
    points = data[['lon','lat']].values.tolist()
    sizes = np.array(data['size'])
    directions = np.array(data['dir'])
    number = np.array(data['num'])
    
    
    for i in range(len(sizes)): 
        try: 
            folium.PolyLine([points[i],points[i+1]],
                            weight=setWeight(sizes[i]),
                            opacity=0.8,
                            color=setColour(directions[i])
                           ).add_to(m)
        except:
            print("")
            
city_markers = folium.FeatureGroup(name='Cities')

for i in range(0,len(cities)):
    latitude = cities['lat'][i]
    longitude = cities['lon'][i]
    name = cities['name'][i]

    popup = folium.Popup(name, parse_html=True) #Create a popup
    marker = folium.Marker([latitude, longitude], popup=popup) #Create a marker
    marker.add_to(city_markers)
    #m.add_child(marker)
    
city_markers.add_to(m)

temp_markers = folium.FeatureGroup(name='Temperature')

for i in range(0,len(temps)):
    latitude = temps['lat'][i]
    longitude = temps['lon'][i]
    name = temps['temp'][i]

    popup = folium.Popup(str(name) + u'\N{DEGREE SIGN}' + 'C', parse_html=True) #Create a popup
    icon=folium.Icon( color='red')
    marker = folium.Marker([latitude, longitude], popup=popup,icon=icon) #Create a marker
    marker.add_to(temp_markers)

temp_markers.add_to(m)
    
folium.LayerControl(collapsed=False).add_to(m)
m.save('napoleons.html')
m
```
